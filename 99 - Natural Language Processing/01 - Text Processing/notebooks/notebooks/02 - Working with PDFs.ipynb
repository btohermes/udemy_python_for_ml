## Notas:
- É importante saber que nem todos os ficheiros de pdf têm texto que pode ser extraído.
- Alguns pdfs são criados através de scanning, em vez de serem exportados de um ficheiro de word. No processo de scan existe uma forte chance serem escritos como imagens, dificultando imenso o procesos de extração dos ficheiros.
- Requer software especializado (aka - Pago !).

```python
import PyPDF2
```

```python
myfile = open("../data/US_Declaration.pdf", mode="rb")
```

```python
pdf_reader = PyPDF2.PdfReader(myfile)
len(pdf_reader.pages)
```

```python
pdf_reader.pages[0].extract_text()
```

## Observação:
No texto acima podes reparar que existe "\n", o código para de nova linha para o processador.
Usar o print para mostrar o texto com as novas linhas associadas.

```python
print(pdf_reader.pages[0].extract_text())
```

```python
f = open("../data/US_Declaration.pdf", mode="rb")
pdf_reader = PyPDF2.PdfReader(f)
```

```python
first_page = pdf_reader.pages[0]
pdf_writer = PyPDF2.PdfWriter()
pdf_writer.add_page(first_page)
```

```python
pdf_output = open("my_brand_new.pdf","wb")
pdf_writer.write(pdf_output)
```

```python
pdf_output.close()
f.close()
```
